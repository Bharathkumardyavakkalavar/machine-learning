```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv(r'C:\Users\bhara\OneDrive\Desktop\all about ml\CODES\student_clustering.csv')

plt.scatter(data['cgpa'], data['iq'])
data.head()
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(data)
    wcss.append(km.inertia_)
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
km=KMeans(n_clusters=4)
x=data.iloc[:,:].values
ans=km.fit_predict(x)
f=x[ans==3]
g=x[ans==2]
h=x[ans==1]
i=x[ans==0]
plt.scatter(f[:,0],f[:,1],color='r')
plt.scatter(g[:,0],g[:,1],color='b')
plt.scatter(h[:,0],h[:,1],color='c')
plt.scatter(i[:,0],i[:,1],color='y')


```
