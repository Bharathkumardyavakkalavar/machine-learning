```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_repeated=0, n_classes=2, 
                        n_clusters_per_class=1,weights=None, flip_y=0.01, class_sep=1.0, hypercube=True,
                        shift=0.0, scale=1.0, shuffle=True, random_state=41)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize =(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
```

```python
def step(l):
    if (l<0): return(0)
    if(l>=0): return(1)    
def percept(x,y):
    x=np.insert(x,0,1,axis=1)
    w=np.ones(x.shape[1])
    lr=0.1
    for i in range (1000000):
        j=np.random.randint(0,100)
        yh=step(np.dot(x[j],w))
        w=w+lr*(y[j]-yh)*x[j]
    return(w)    
w=percept(x,y)   
```

```python
def predict(p):
    return np.dot(p,w)
d=0
f=-2  
print(predict([1,d,f])>=0 )   
```
